```python
texts = []
for page in pages:
    text = pytesseract.image_to_string(Image.open(page).convert('L'))
    print(page)
    texts.append(text)
    
```

```python
import cv2 as cv
import PIL
from PIL import Image
import numpy as np
from PIL import ImageDraw
from math import ceil

def find_faces(image_name):
    img = cv.imread(image_name)
    image = Image.open(image_name)
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    
    try:
        faces = face_cascade.detectMultiScale(gray,1.35).tolist()
    except:
        print('But there were no faces in that file!!!')
        return
    
    pil_img=Image.fromarray(gray,mode="L")
    drawing=ImageDraw.Draw(pil_img)
    
    maxW = 0
    maxH = 0 
    
    n = len(faces)
    if n == 0:
        print('But there were no faces in that file!!!')
        return
    
    for x,y,w,h in faces:
        if w>maxW:
            maxW = w
        if h>maxH:
            maxH = h
    
    contact_sheet=PIL.Image.new(pil_img.mode, (maxW*5,maxH*ceil(n/5)))
    i=0
    j=0

    for x,y,w,h in faces:
        cropped = pil_img.crop((x,y,x+w,y+h))
        #display(cropped)

        contact_sheet.paste(cropped, (i, j) )

        if i+maxW == contact_sheet.width:
            i=0
            j=j+maxH
        else:
            i=i+maxW
    
    
    display(contact_sheet)
    #for rec in faces:
        #print(rec)
    return
    
```

```python
pages = ['a-0.png','a-1.png','a-2.png','a-3.png']
for text in texts[0:4]:
    i = texts.index(text)
    if "Christopher" in text:
        print("Results found in a-%.f.png" %i )  #so far so good
        find_faces(pages[i])
        
        
```
